write students data in csv

```python
import csv

def write_student_data():
    """Writes student data to students.csv file."""
    
    # Data to write
    students = [
        ["Name", "Age", "Grade"],
        ["Rahul", 16, "10th"],
        ["Sakshi", 15, "9th"],
        ["Aman", 17, "11th"]
    ]

    # Writing to CSV
    with open("students.csv", "w", newline="") as file:
        writer = csv.writer(file)
        writer.writerows(students)

    print("Student data written to students.csv")


# ---- Call the function ----
write_student_data()
```

```python
## read the csv file(students.csv)

import csv

def read_student_data():
    """Reads and prints data from students.csv"""
    with open("students.csv", "r") as file:
        reader = csv.reader(file)
        for row in reader:
            print(row)
read_student_data()
```

```python
## append in student.csv file

import csv

def append_student_data():
    """Appends one more student record to students.csv"""
    
    # New student to append
    new_student = ["Riya", 16, "10th"]

    with open("students.csv", "a", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(new_student)

    print("New student added successfully!")

# Call the function
append_student_data()
```
